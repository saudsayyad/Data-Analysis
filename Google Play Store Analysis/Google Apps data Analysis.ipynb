```python
from sklearn.preprocessing import LabelEncoder

# Initialize Label Encoder
label_encoder = LabelEncoder()
df2 = df[['Category','Genres']]
# Apply Label Encoding to each column
df2['Category_encoded'] = label_encoder.fit_transform(df2['Category'])
df2['Genres_encoded'] = label_encoder.fit_transform(df2['Genres'])

corr = df2['Category_encoded'].corr(df2['Genres_encoded'])
print("Correlation:",corr)
```

```python
final_result = df.sort_values(by = 'Installs', ascending = False).head(10)
plt.bar("App", 'Installs', data = final_result, color = "blue")
plt.xlabel("Apps")
plt.xticks(rotation = 90)
plt.ylabel("Installs")
plt.show()
```

```python
final_result = df.sort_values(by = 'Size', ascending = False).head(10)
plt.bar("App", 'Size', data = final_result, color = "blue")
plt.xlabel("Apps")
plt.xticks(rotation = 90)
plt.ylabel("Size")
plt.show()
```
